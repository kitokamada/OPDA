```python
# Header Fiding Rrs at Available lambda Bands 
header_array=row_arrays[3]
data_string = header_array[0]
# Split the string by commas to get individual elements
all_headers = data_string.split(',')
all_headers=np.array(all_headers)
```

```python
target_headers = np.array(['env_rrs412', 'env_rrs443', 'env_rrs490', 'env_rrs510', 'env_rrs560','env_rrs620','env_rrs667','env_rrs681','env_rrs709'])
# Find indices of target Rrs (column)
Rrs_indices = [np.where(all_headers == header)[0] for header in target_headers]
Rrs_indices = [elem[0] for elem in Rrs_indices if elem.size > 0]
```

```python
split_rows = np.array([row[0].split(',') for row in row_arrays][4:])
second_elements = np.array([row[1] for row in split_rows if len(row) > 1])  # this arrary contains all the station names in sequence of measurement of the csv file on PACE website
# Revise Station Name if needed
target_station=np.array(['Casablanca_Platform'])
# Find indices of target station (row)
target_station_indices = [np.where(second_elements == station) for station in target_station]
target_rows=np.squeeze(split_rows[target_station_indices])
Rrs_Value_AERONET=np.array(target_rows[:, Rrs_indices]).astype(float)
```
